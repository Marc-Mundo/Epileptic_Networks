```python
def compute_manual_firing_rate(spike_times, stim_duration, dt):
    spike_times_list = list(spike_times)
    num_spikes = len(spike_times_list)
    total_time = (
        len(np.arange(0, stim_duration, dt)) * dt / 1000
    )  # Convert stim_duration and dt to seconds
    firing_rate = num_spikes / total_time
    return firing_rate


# Flatten and concatenate spike times
pyr_spike_times = np.concatenate([cell.spike_times for cell in pyr_cells])
bwb_spike_times = np.concatenate([cell.spike_times for cell in bwb_cells])
olm_spike_times = np.concatenate([cell.spike_times for cell in olm_cells])

# Simulation parameters
simulation_duration = 5000  # in milliseconds
dt = 0.1  # time step in milliseconds

# Compute firing rates for each cell and population
pyr_firing_rates = [
    compute_manual_firing_rate(cell.spike_times, simulation_duration, dt)
    for cell in pyr_cells
]
bwb_firing_rates = [
    compute_manual_firing_rate(cell.spike_times, simulation_duration, dt)
    for cell in bwb_cells
]
olm_firing_rates = [
    compute_manual_firing_rate(cell.spike_times, simulation_duration, dt)
    for cell in olm_cells
]

# Create lists for populations
pyr_mean, pyr_std = np.mean(pyr_firing_rates), np.std(pyr_firing_rates)
bwb_mean, bwb_std = np.mean(bwb_firing_rates), np.std(bwb_firing_rates)
olm_mean, olm_std = np.mean(olm_firing_rates), np.std(olm_firing_rates)

# Print results
print(f"Pyr :: {pyr_mean:.2f} Hz +- {pyr_std:.2f} Hz (std)")
print(f"Bwb :: {bwb_mean:.2f} Hz +- {bwb_std:.2f} Hz (std)")
print(f"Olm :: {olm_mean:.2f} Hz +- {olm_std:.2f} Hz (std)")
```

```python
# Calculate the lfp based on cell.Adend3_v and cell.Bdend_v
lfp_list = []  # Initialize an empty list to store lfp values

# Assuming Adend3_v and Bdend_v are numpy arrays
for cell in pyr_cells:
    vlfp = h.Vector()
    vlfp.from_python(cell.Adend3_v)
    vlfp.sub(h.Vector().from_python(cell.Bdend_v))
    vlfp.div(len(pyr_cells))
    lfp_list.append(np.array(vlfp.to_python()))

lfp = np.mean(lfp_list, axis=0)  # Calculate the mean of the lfp values along axis 0

print(lfp)
```

```python
from scipy.signal import welch


def calc_psd(lfp, fs=1000, tr=(3, 12), gr=(30, 80)):
    """
    Calculate the mean theta and gamma power of the LFP signal.

    Parameters:
    - lfp: numpy array, LFP signal
    - fs: int, sampling frequency (default is 1000 Hz)
    - tr: tuple, theta frequency range (default is (3, 12) Hz)
    - gr: tuple, gamma frequency range (default is (30, 80) Hz)

    Returns:
    - mean_theta_power: float, mean power in the theta frequency range (sq-mV/Hz)
    - mean_gamma_power: float, mean power in the gamma frequency range (sq-mV/Hz)
    - theta_freq: float, dominant frequency in the theta range (Hz)
    - gamma_freq: float, dominant frequency in the gamma range (Hz)
    """

    # Calculate the power spectral density (PSD) using Welch method
    f, Pxx = welch(lfp, fs=fs, nperseg=1024)

    # Find indices corresponding to theta and gamma frequency ranges
    theta_indices = np.where((f >= tr[0]) & (f <= tr[1]))[0]
    gamma_indices = np.where((f >= gr[0]) & (f <= gr[1]))[0]

    # Calculate mean power within theta and gamma frequency ranges
    mean_theta_power = np.mean(Pxx[theta_indices])
    mean_gamma_power = np.mean(Pxx[gamma_indices])

    # Find dominant frequencies in theta and gamma ranges
    theta_freq = f[theta_indices[np.argmax(Pxx[theta_indices])]]
    gamma_freq = f[gamma_indices[np.argmax(Pxx[gamma_indices])]]

    print(f"Mean Theta Power: {mean_theta_power:.3e} sq-mV/Hz")
    print(f"Mean Gamma Power: {mean_gamma_power:.3e} sq-mV/Hz")
    print(f"Dominant Theta Frequency: {theta_freq:.2f} Hz")
    print(f"Dominant Gamma Frequency: {gamma_freq:.2f} Hz")

    return mean_theta_power, mean_gamma_power, theta_freq, gamma_freq


# Example usage:
mean_theta_power, mean_gamma_power, theta_freq, gamma_freq = calc_psd(lfp)
```
