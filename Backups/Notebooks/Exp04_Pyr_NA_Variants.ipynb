```python
# Import load function
from src.SanjayCode import load

# Path to data directory
data_path = "../data/Data05_Pyr_NA_Variants/"

# Define the runs to be analyzed and the corresponding labels
my_runs = [
    "NA_0.5",
    "NA_0.6",
    "NA_0.7",
    "NA_0.8",
    "NA_0.9",
    "NA_1.0",
    "NA_1.1",
    "NA_1.2",
    "NA_1.3",
    "NA_1.4",
    "NA_1.5",
]

# Load the data
data = load(data_path)
```

# Process data


```python
# Import process_data function
from src.SanjayCode import process_data

# Dictionary to store results for each condition and run
results = {}

# Iterate over each condition and run
for condition, runs in data.items():
    results[condition] = {}
    for run, simData in runs.items():
        print(f"Processing {condition}, run {run}")
        results[condition][run] = process_data(simData["simData"])
```
